# findpeaks Python package 
find more information on the findpeaks package on
[https://erdogant.github.io/findpeaks/pages/html/Abstract.html](https://erdogant.github.io/findpeaks/pages/html/Abstract.html)

```python
# Import library
from findpeaks import findpeaks
import numpy as np
import demo_data
```

```python
# Initialize
fp = findpeaks(lookahead=1)
x, y = demo_data.coelute()
results = fp.fit(y)
# Plot
fp.plot()
```

```python
# Initialize
fp = findpeaks(method='peakdetect', lookahead=5, interpolate=None)

x, y = demo_data.sin_peaks(100)

# Fit topology method on the 1d-vector
results = fp.fit(y)
# Plot
fp.plot()
```

```python
# Initialize
fp = findpeaks(method='peakdetect', lookahead=5, interpolate=None)

x, y = demo_data.chromatogram(noise_stddev=0.06, data_points=200)

# Fit topology method on the 1d-vector
results = fp.fit(y)
# Plot
fp.plot()
```
