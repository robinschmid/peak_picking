# Example using scipy.find_peaks, Savitzky Golay smoothing, and CWT

```python
import numpy as np
from scipy.signal import find_peaks, savgol_filter, find_peaks_cwt
import matplotlib.pyplot as plt
import demo_data
from plotting import scipy_plotting_smoothed, scipy_plotting
```

```python
xy_data = demo_data.noisy_sin()
scipy_plotting_smoothed(xy_data, height=0.5, distance=50, smooth_window_length=50, smooth_polyorder=3)
```

```python
xy_data = demo_data.chromatogram(noise_stddev=0.06, data_points=200)
scipy_plotting_smoothed(xy_data, height=0.3, distance=10, smooth_window_length=5, smooth_polyorder=3)
```

# Continuous Wavelet Transform (CWT)

```python
# Apply Continuous Wavelet Transform (CWT)
x, y = demo_data.chromatogram(noise_stddev=0.06, data_points=200)
widths = np.arange(2, 10)  # Adjust the range of widths as needed
peaks_indices = find_peaks_cwt(y, widths)
scipy_plotting(x, y, peaks_indices)
```
